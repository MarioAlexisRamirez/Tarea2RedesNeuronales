```python
#EJERCICIO 1, INCISO a)
#importar las librerías de python



import numpy as np
import re
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from matplotlib import pyplot
from keras.layers import Dense

np.set_printoptions(suppress=True)

x = asarray([i/500 for i in range(-500,500)])
y = asarray([3*np.sin(np.pi*i)  for i in x])

x = x.reshape((len(x), 1))
y = y.reshape((len(y), 1))

scale_x = MinMaxScaler()
x = scale_x.fit_transform(x)
scale_y = MinMaxScaler()
y = scale_y.fit_transform(y)

model = Sequential()
model.add(Dense(25, input_dim=1, activation='tanh', kernel_initializer='he_uniform'))        #con otras funciones de activación no funciona bien
model.add(Dense(20, input_dim=1, activation='sigmoid', kernel_initializer='he_uniform'))
model.add(Dense(20, input_dim=1, activation='sigmoid', kernel_initializer='he_uniform'))
model.add(Dense(1))

model.compile(loss='mse', optimizer='RMSprop')
model.fit(x, y, epochs=40, batch_size=8, verbose=1)
yh = model.predict(x)

# escalar los ejes



x_plot = scale_x.inverse_transform(x)
y_plot = scale_y.inverse_transform(y)
yh_plot = scale_y.inverse_transform(yh)

# gráficas juntas en un solo plano.



pyplot.scatter(x_plot,y_plot, color='red', label='Solución real')
pyplot.scatter(x_plot,yh_plot,color='darkgreen', label='Solución predicha')
pyplot.plot(x_plot,yh_plot, "--")
pyplot.title(r'$Gráficas real y predicha$')
pyplot.grid()
pyplot.legend()
pyplot.figure(figsize=(15,8))
pyplot.show()


```

```python
#EJERCICIO 1, INCISO b)


#importar las librerías de python
import numpy as np
import re
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from matplotlib import pyplot
from keras.layers import Dense
np.set_printoptions(suppress=True)

x = asarray([i/500 for i in range(-500,500)])
y = asarray([4*np.power(i,3) + (2*i) + 1  for i in x])

x = x.reshape((len(x), 1))
y = y.reshape((len(y), 1))

scale_x = MinMaxScaler()
x = scale_x.fit_transform(x)
scale_y = MinMaxScaler()
y = scale_y.fit_transform(y)

model = Sequential()
model.add(Dense(30, input_dim=1, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(20, input_dim=1, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1))

model.compile(loss='mse', optimizer='RMSprop')
model.fit(x, y, epochs=40, batch_size=8, verbose=1)
yh = model.predict(x)

#escalas de los ejes


x_plot = scale_x.inverse_transform(x)
y_plot = scale_y.inverse_transform(y)
yh_plot = scale_y.inverse_transform(yh)

# graficas con matplotlib las dos gráficas deben aparecer en un solo plano.


pyplot.scatter(x_plot,y_plot, color='purple',  label='Solución real')
pyplot.scatter(x_plot,yh_plot, color='darkorange', label='Solución predicha')
pyplot.plot(x_plot,yh_plot, "--")
pyplot.title(r'Gráficas real y predicha')
pyplot.grid()
pyplot.legend()
pyplot.figure(figsize=(15,8))
pyplot.show()
```
